# Temperature rise

Measurement date: August 4, 2018

Location: second floor of my house

Heat the following materials under strong sunshine:
- Black metal plate
- White metal plate
- Black sponge rubber

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.signal as sg

matplotlib.rcParams['figure.figsize'] = [16, 7]
FILE = 'temperature.csv'
```

```python
df = pd.read_csv(FILE)
df.head()
ax = df[['white metal', 'black metal','sponge rubber']].plot(x=df['elapsed time'], title='Temperature')
ax.set_xlabel('Elaplsed time in sec')
ax.set_ylabel('Temperature in degrees Celsius')
```
